### 添加椒盐噪声

```python
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt 
def noiseSP(img, prob):
    '''
    添加椒盐噪声
    :param img: 原始图片
    :param prob: 噪声比例
    :return: resultImg 这是一个张量矩阵
    '''
    resultImg = np.zeros(img.shape, np.uint8)#初始化大小为图像大小的矩阵（张量），uint8是专门用于存储各种图像的（包括RGB，灰度图像等），范围是从0–255
    thres = 1 - prob
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            rdn = random.random()  # 随机生成0-1之间的数字
            if rdn < prob:  # 如果生成的随机数小于噪声比例则将该像素点添加黑点，即椒噪声
                resultImg[i][j] = 0
            elif rdn > thres: # 如果生成的随机数大于（1-噪声比例）则将该像素点添加白点，即盐噪声
                resultImg[i][j] = 255
            else:
                resultImg[i][j] = img[i][j]  
    return resultImg
 
img = cv2.imread('./lenda.png')
resultImg = noiseSP(img, 0.05)
fig,[axes1,axes2]=plt.subplots(1,2,figsize=(10,10)) #生成画布
axes1.imshow(img)
axes2.imshow(resultImg)
```
