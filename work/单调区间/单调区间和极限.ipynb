```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

## 求单调区间

#### 画图观察y = x**3-3*x 

```python
x = np.linspace(-3,3,100)
y = x**3-3*x 
plt.plot(x,y)
plt.grid(True)
plt.show()
```

#### 求一阶导数，0点

```python
x = symbols('x')
y = x**3-3*x 
ds_1 = diff(y,x)  #diff是求导数
ans = solve(ds_1,x)
print('函数的导数为',ds_1)
print('1阶导数为0的x点是',ans)
```

#### 通过上述可知0点分别为x=-1和x=1 (下面画了3个区间(-3,-1)(-1,1)(1,3))

```python
#显示中文
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

#### 通过下面画图  (-∞,-1)单调递增 (-1,1)单调递减"(1,+∞)单调递增

```python
def fun(x):
    return x**3-3*x 
fig,axes=plt.subplots()
x1 = x = np.linspace(-3,-1,100)
x2 = x = np.linspace(-1,1,100)
x3 = x = np.linspace(1,3,100)
axes.plot(x1,fun(x1))
axes.plot(x2,fun(x2))
axes.plot(x3,fun(x3))
axes.legend(["单调递增", "单调递减","单调递增"], loc=0) 
plt.grid(True)
plt.show()
```

## 求极限

#### 首先观察图像

```python
def fun(x):
    return np.sin(x)/x 
x1=np.linspace(-6*np.pi,6*np.pi,100)
plt.plot(x1,fun(x1))
plt.grid(True)
plt.show()
```

#### 求sin(x)/x的任一点极限

```python
def my_lim(x0):
    if x0==0: 
        x=Symbol("x")
        return limit(sin(x)/x, x, x0)
    else:
        return np.sin(x0)/x0
```

#### x0=0的极限

```python
my_lim(0)
```

#### 求任一点的极限

```python
my_lim(2)
```
