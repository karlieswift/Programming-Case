* https://data.worldbank.org.cn/indicator/NY.GDP.MKTP.CD

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

### 面积

```python
data=pd.read_excel('./Area_Data_Extract_From_世界发展指标.xlsx')
data
```

#### 国家面积排名

```python
data_1=data.iloc[:,[2,-3]]
data_1.columns=["国家","2018面积"]
data_1
```

```python
data_1=data_1.sort_values(by="2018面积",ascending=False)
data_1.head()
```

```python

plt.title("面积排名")
sns.barplot(y="国家",x="2018面积",data=data_1.head(),orient='h')
```

```python
data_1.to_excel('./面积排名.xls')
```

#### 面积对比

```python
def pie_cla(data_1):
    total=data_1['2018面积'].sum()
    a=data_1.iloc[0,1]
    b=data_1.iloc[1,1]
    c=data_1.iloc[2,1]
    d=data_1.iloc[3,1]
    other=total-a-b-c-d 
    return [a,b,c,d,other]
pie_cl=pie_cla(data_1)
label = ['俄罗斯联邦','加拿大','美国','中国',"其他"] 
plt.figure(figsize=(10,6)) 
plt.title('top5国土面积占比')
explode = (0, 0,0,0,0) 
plt.pie(pie_cl, explode=explode,labels=label, autopct='%1.1f%%')#autopct保留位数
```

#### 国家面积变化

```python
a=[]
for i in range(len(data)):
    v=np.var(data.iloc[i,5:])
    a.append([data.iloc[i,2],v])
county_Area_var=pd.DataFrame(a,columns=["国家","var"])
county_Area_var=county_Area_var.sort_values(by="var",ascending=False)
county_Area_var.to_excel('./面积方差.xls')
```
